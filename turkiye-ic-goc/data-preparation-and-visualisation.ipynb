```python
df = df.rename(columns=normalized_column_names)
df.head()
```

```python
df.columns[3:-3]
```

```python
df_sebepler = df[df['Flag']=='Al'].groupby('Yil')[df.columns[3:-3]].sum()
df_sebepler.head()
```

```python
kolon_sirali = df_sebepler.sum().sort_values(ascending=False).index
df_sebepler_sirali = df_sebepler[kolon_sirali]
df_sebepler_sirali.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,8))
for column in df_sebepler_sirali.columns[0:5]:
  plt.plot(df_sebepler_sirali.index,df_sebepler_sirali[column],label=column)

plt.xticks(df_sebepler_sirali.index)
plt.xlabel('Yil')
plt.ylabel('Goc Sebebi')
plt.legend(loc='upper right')
plt.show()
```

```python
df_sehir = df[['Yil','Sehir','Toplam','Flag']]
df_sehir.head()
```

```python
df_sehir_pivot = pd.pivot_table(df_sehir, values='Toplam', index=['Yil','Sehir'],columns=['Flag'])
df_sehir_pivot.reset_index(inplace=True)
df_sehir_pivot.head()
```

```python
df_sehir_pivot['Delta']=df_sehir_pivot['Al']-df_sehir_pivot['Ver']
df_sehir_pivot['Mutlak Delta'] = df_sehir_pivot['Delta'].apply(abs)
df_sehir_pivot
```

```python
df_sehir_grouped = df_sehir_pivot.groupby('Sehir')['Al','Ver','Mutlak Delta','Delta'].sum()
df_sehir_grouped
```

```python
df_sehir_grouped_top_10 = df_sehir_grouped.sort_values(by='Mutlak Delta', ascending=False).head(10)
df_sehir_grouped_top_10.head()
```

```python
df_sehir_grouped_delta_10 = df_sehir_pivot[df_sehir_pivot['Sehir'].isin(list(df_sehir_grouped_top_10.index))]
df_sehir_grouped_delta_10.head
```

```python
df_sehir_grouped_delta_10_pivot = df_sehir_grouped_delta_10[df_sehir_grouped_delta_10['Sehir']!='İstanbul'].pivot(index='Yil', columns='Sehir', values='Delta')

df_sehir_grouped_delta_10_pivot.plot(figsize=(16,8))
plt.xticks(df_sebepler_sirali.index)
plt.xlabel('Yil')
plt.ylabel('Sehir')
plt.legend(loc='upper right')
plt.show()
```
